```python
#Create 2D pdf
from numpy import random, histogram2d, diff
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d

# Generate sample data
n = 10000
x = random.randn(n)
y = -x + random.randn(n)

# bin
nbins = 50
H, xedges, yedges = histogram2d(x, y, bins=nbins,normed=1)

# Figure out centers of bins
def centers(edges):
    return edges[:-1] + diff(edges[:2])/2

xcenters = centers(xedges)
ycenters = centers(yedges)

# Construct interpolator
pdf = interp2d(xcenters, ycenters, H)

H, xedges, yedges = histogram2d(1*x, 1*y, bins=nbins,normed=1)
H=H.T
xcenters = centers(xedges)
ycenters = centers(yedges)

pdf2 = interp2d(xcenters, ycenters, H)

# test
fig=plt.contour(xedges, yedges, pdf(xedges, yedges)*pdf2(xedges, yedges)/(pdf2(xedges, yedges)*pdf2(xedges, yedges)),10)

plt.colorbar()
#plt.colormaps
```
