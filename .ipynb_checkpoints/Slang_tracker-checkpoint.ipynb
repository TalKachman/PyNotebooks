<h1> Clean up data. Remove stop list and web links </h1> 

```python
slang_sentence= [[word for word in sentence[0].lower().split() if word not in stoplist \
                  and re.findall('http',word)==[] and re.findall('@',word)==[]\
                 and re.findall('#',word)==[]] for sentence in results]
```

```python
norm_sentence= [[word for word in sentence[0].lower().split() if word not in stoplist \
                  and re.findall('http',word)==[] and re.findall('@',word)==[]\
                and re.findall('@',word)==[]] for sentence in norm_results]
```

```python
norm_sentence
```

<h1> train model </h1>

```python
model = models.Word2Vec(norm_sentence+slang_sentence, size=100, window=5, min_count=10, workers=4)
```

```python
model.most_similar('bacon')
```

```python
a=model.most_similar('beer')
for val in a:
    print(val[0])
```

<h1> work with bigrams </h1>

```python
bigram = models.Phrases(norm_sentence)
#print(list(bigram[norm_sentence[10:20]]))
bigrams=list(bigram[norm_sentence])
```

```python
modelB = models.Word2Vec(bigrams, size=100, window=5, min_count=10, workers=4)
```

```python
modelB.most_similar(positive=['finna', 'gonna'], negative=['bro'],topn=20)
```
