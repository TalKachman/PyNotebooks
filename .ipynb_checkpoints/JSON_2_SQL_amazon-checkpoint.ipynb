<h1> Convert json text to SQL DB </h1>

```python
from __future__ import division
import psycopg2
import numpy as np
import json
import re
from ast import literal_eval
```

```python
#conn = psycopg2.connect("host=localhost port=5432 dbname=qa")
conn = psycopg2.connect("host=localhost port=5432 dbname=qa")
cur = conn.cursor()
```

<h1> Create SQL table to store the data </h1>

```python
#QA db
cur.execute("CREATE TABLE qa (id serial PRIMARY KEY,\
    catagory text,\
    questionType text,\
    answerType text,\
    asin text,\
    answerTime TimeStamp,\
    unixTime int,\
    question text,\
    answer text\
    );")
conn.commit() #submit change to db
```

<h1> Def function to insert values into postgres db </h1>

```python
def writeJSON2SQL(catagory,questionType,answerType,asin,answerTime,unixTime,question,answer):
    if answerTime=='':
        answerTime=None
    try:
        unixTime=int(unixTime)
    except ValueError:
        unixTime=0
    
    cur.execute("INSERT INTO qa (\
    catagory,\
    questionType,\
    answerType,\
    asin,\
    answerTime,\
    unixTime,\
    question,\
    answer\
    )\
    VALUES (\
    %s,%s,%s,%s,%s,%s,%s,%s\
    )",(catagory,questionType,answerType,asin,answerTime,unixTime,question,answer)\
    )
    conn.commit() #submit change to db
```

<h1> Def function to read in the data </h1>

```python
def read_line(line):
    questionType=re.findall(r'questionType\': \'(.*?)\',',line)
    if questionType==[]:
        questionType=['']
    answerType=re.findall(r'answerType\': \'(.*?)\',',line)
    if answerType==[]:
        answerType=['']
    asin=re.findall(r'asin\': \'(.*?)\',',line)
    answerTime=re.findall(r'answerTime\': \'(.*?)\',',line)
    if answerTime==[]:
        answerTime=['']
    if answerTime[0].count(" ago")>0:
        answerTime=['']
    unixTime=re.findall(r'unixTime\': (.*?),',line)
    if unixTime==[]:
        unixTime=['']
    question=re.findall(r'question\': \'(.*?)\',',line)
    if question==[]:
        question=re.findall(r'question\': "(.*?)"',line)
    answer=re.findall(r'answer\': \'(.*?)\'',line)
    if answer==[]:
        answer=re.findall(r'answer\': "(.*?)"',line)
    return questionType[0],answerType[0],asin[0],answerTime[0],unixTime[0],question[0],answer[0]
```

<h1> Read all the QA files from Tall Idea Lab and write to SQL</h1>

```python
file_list=!ls /Users/attiladobi/Downloads/QA
for catagory_file in file_list:
    lines=open('/Users/attiladobi/Downloads/QA/'+catagory_file).read().split('\n')[:-1]
    for ii,line in enumerate(lines):
        questionType,answerType,asin,answerTime,unixTime,question,answer = read_line(line)
        writeJSON2SQL(catagory_file,questionType,answerType,asin,answerTime,unixTime,question,answer)
```

<h3> create the table reviews</h3>

```python
#QA db
cur.execute("CREATE TABLE reviews (id serial PRIMARY KEY,\
    asin text,\
    helpful int[],\
    overall float,\
    reviewText text,\
    reviewTime TimeStamp,\
    reviewerID text,\
    reviewerName text,\
    summary text,\
    unixReviewTime int\
    );")
conn.commit() #submit change to db
```

<h3> loop through data and dump into the reviews table </h3>

```python
for entry in data:
    asin=entry['asin']
    helpful=entry['helpful']
    overall=entry['overall']
    reviewText=entry['reviewText']
    reviewTime=entry['reviewTime']
    reviewerID=entry['reviewerID']
    try:
        reviewerName=entry['reviewerName']
    except KeyError:
        reviewerName=''
    summary=entry['summary']
    unixReviewTime=entry['unixReviewTime']
        
    cur.execute("INSERT INTO reviews (\
    asin,\
    helpful,\
    overall,\
    reviewText,\
    reviewTime,\
    reviewerID,\
    reviewerName,\
    summary,\
    unixReviewTime\
    )\
    VALUES (\
    %s,%s,%s,%s,%s,%s,%s,%s,%s\
    )",(asin,helpful,overall,reviewText,reviewTime,reviewerID,reviewerName,summary,unixReviewTime)\
    )
    conn.commit() #submit change to db
```
